```python
# Getting start dates and end dates of weeks 1-4 for all 10 courses
from datetime import timedelta 
start_dates = []
end_dates = []
n = len(courses)
for i in range(n):
    start_temp = []
    end_temp = []
    c = courses.iloc[i]
    start_date = pd.to_datetime(c['from'], infer_datetime_format=True)
    for j in range(4):
        sd = str(start_date).split(" ")[0]
        start_temp.append(sd)
        end_date = start_date + pd.Timedelta(timedelta(days = 6))
        ed = str(end_date).split(" ")[0]
        end_temp.append(ed)
        start_date = end_date + pd.Timedelta(timedelta(days = 1))
    end_date = pd.to_datetime(c['to'], infer_datetime_format=True)
    ed = str(end_date).split(" ")[0]
    end_temp[-1] = ed
    start_dates.append(start_temp)
    end_dates.append(end_temp)
```

```python
course_id_list = courses['course_id'].to_list()
dates_dict = {'course_id':course_id_list}
start_transpose = np.array(start_dates).T.tolist()
end_transpose = np.array(end_dates).T.tolist()
for i in range(4):
    start_colname = "week"+str(i+1)+"_start"
    dates_dict[start_colname] = start_transpose[i]
    end_colname = "week"+str(i+1)+"_end"
    dates_dict[end_colname] = end_transpose[i]
```

```python
start_end_dates = pd.DataFrame(dates_dict)
# saving the dataframe
start_end_dates.to_csv('start_end_dates.csv')
```
