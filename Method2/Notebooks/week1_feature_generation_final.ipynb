```python
import pandas as pd
import os
import numpy as np
import datetime
from datetime import date
```

```python
truth_train = pd.read_csv("train/truth_train.csv")
truth_enroll = truth_train['1'].to_list()
```

```python
required_train_users = pd.read_csv("required_users.csv")
required_train_users.drop('Unnamed: 0', axis=1, inplace=True)
#required_train_users = required_train_users.head(10000) # Change the number as per requirement
```

```python
required_train_users
```

```python
enrollment_id_list = required_train_users['enrollment_id'].to_list()
```

```python
week1 = pd.read_csv('week1.csv')
week1
```

```python
week1.drop('Unnamed: 0', axis=1, inplace=True)
week1
```

```python
features1 = []
enrollment_id1 = []
week1_len = len(week1)
for i in range(week1_len):
    e_id = week1.loc[i, 'enrollment_id']
    if e_id not in enrollment_id1:
        enrollment_id1.append(e_id)
#enrollment_id1 = enrollment_id1[:2]
for eid in enrollment_id1:
    i = enrollment_id_list.index(eid)
    truth_index = truth_enroll.index(eid)
    temp1 = [eid, required_train_users.loc[i, 'username'], required_train_users.loc[i, 'course_id']]+[0]*23+[truth_train.loc[truth_index, '0']]
    features1.append(temp1)
unique_enroll = len(enrollment_id1)
```

```python
unique_enroll
```

```python
# Feature 1 - Number of participating objects of course per week
objs = {}
for i in range(week1_len):
    ob = week1.loc[i, 'object']
    eid = week1.loc[i, 'enrollment_id']
    index = enrollment_id1.index(eid)
    if eid not in objs:
        objs[eid] = [ob]
        features1[index][3] += 1
    else:
        if ob not in objs[eid]:
            objs[eid].append(ob)
            features1[index][3] += 1
```

```python
# Feature 2-8 - Behaviour number of access, page_close, problem video, discussion, navigate, wiki
for i in range(week1_len):
    eid = week1.loc[i, 'enrollment_id']
    event = week1.loc[i, 'event']
    index = enrollment_id1.index(eid)
    if event == 'access':
        features1[index][4] += 1
    elif event == 'page_close':
        features1[index][5] += 1
    elif event == 'problem':
        features1[index][6] += 1
    elif event == 'video':
        features1[index][7] += 1
    elif event == 'discussion':
        features1[index][8] += 1
    elif event == 'navigate':
        features1[index][9] += 1
    else:
        features1[index][10] += 1  # wiki
```

```python
# Feature 9-10 - The total and average numbers of all behaviors per week
for i in range(unique_enroll):
    behaviour_list = features1[i][4:11]
    features1[i][11] = sum(behaviour_list) # Total behaviours per week
    features1[i][12] = features1[i][11]/7   # Average behaviours per week
```

```python
# Feature 11 - Number of active days per week
active_dates = {}
for i in range(week1_len):
    d = week1.loc[i, 'date']
    eid = week1.loc[i, 'enrollment_id']
    index = enrollment_id1.index(eid)
    if eid not in active_dates:
        active_dates[eid] = [d]
        features1[index][13] += 1
    else:
        if d not in active_dates[eid]:
            active_dates[eid].append(d)
            features1[index][13] += 1
```

```python
# Feature 12 - Time consumption per week
t1 = pd.to_datetime(week1.loc[0, 'time'], infer_datetime_format=True)
e1 = week1.loc[0, 'enrollment_id']
d1 = week1.loc[0, 'date']
for i in range(1, week1_len):
    t2 = pd.to_datetime(week1.loc[i, 'time'], infer_datetime_format=True)
    e2 = week1.loc[1, 'enrollment_id']
    d2 = week1.loc[1, 'date']
    index = enrollment_id1.index(e2)
    if e1==e2 and d1==d2:
        delta = (t2 - t1).total_seconds()
        features1[index][14] += delta
    t1 = t2
    e1 = e2
    d1 = d2    
```

```python
# Feature 13-16 - The behavior numbers of access, page close, problem, video from browser per week respectively.
for i in range(week1_len):
    eid = week1.loc[i, 'enrollment_id']
    source = week1.loc[i, 'source']
    event = week1.loc[i, 'event']
    index = enrollment_id1.index(eid)
    if source=='browser':
        if event == 'access':
            features1[index][15] += 1
        elif event == 'page_close':
            features1[index][16] += 1
        elif event == 'problem':
            features1[index][17] += 1
        elif event == 'video':
            features1[index][18] += 1
```

```python
# Feature 17-21 - The behavior numbers of access, discussion, problem, navigating, wiki from server per week respectively
for i in range(week1_len):
    eid = week1.loc[i, 'enrollment_id']
    source = week1.loc[i, 'source']
    event = week1.loc[i, 'event']
    index = enrollment_id1.index(eid)
    if source=='server':
        if event == 'access':
            features1[index][19] += 1
        elif event == 'discussion':
            features1[index][20] += 1
        elif event == 'problem':
            features1[index][21] += 1
        elif event == 'navigate':
            features1[index][22] += 1
        elif event == 'wiki':
            features1[index][23] += 1  # wiki
```

```python
# Feature 22-23 - The numbers of all behaviors from browser and server per week respectively
for i in range(unique_enroll):
    browser_list = features1[i][15:19]
    features1[i][24] = sum(browser_list) # Total behaviours from browser per week
    server_list = features1[i][19:24]
    features1[i][25] = sum(server_list) # Total behaviours from server per week
```

```python
cols = ['enrollment_id',
        'username',
        'course_id',
        'no_of_objects1',
        'bno_access1',
        'bno_pageclose1',
        'bno_problem1',
        'bno_video1',
        'bno_discussion1',
        'bno_navigate1',
        'bno_wiki1',
        'bno_total1',
        'bno_avg1',
        'no_of_active_days1',
        'time_consumption1',
        'browser_access1',
        'browser_pageclose1',
        'browser_problem1',
        'browser_video1',
        'server_access1',
        'server_discussion1',
        'server_problem1',
        'server_navigate1',
        'server_wiki1',
        'bno_browser1',
        'bno_server1',
        'label']
        
features_week1 = pd.DataFrame(features1, columns = cols)
```
