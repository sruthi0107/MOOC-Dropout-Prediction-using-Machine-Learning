```python
req_train = pd.read_csv("req_train.csv")
req_train.drop('Unnamed: 0', axis=1, inplace=True)
req_train
```

```python
features = []
enrollment_id1 = []
req_len = len(req_train)
for i in range(req_len):
    e_id = req_train.loc[i, 'enrollment_id']
    if e_id not in enrollment_id1:
        enrollment_id1.append(e_id)
#enrollment_id1 = enrollment_id1[:2]
for eid in enrollment_id1:
    i = enrollment_id_list.index(eid)
    truth_index = truth_enroll.index(eid)
    temp = [eid, required_train_users.loc[i, 'username'], required_train_users.loc[i, 'course_id']]+[0]*23+[truth_train.loc[truth_index, '0']]
    features.append(temp)
unique_enroll = len(enrollment_id1)
```

```python
unique_enroll
```

```python
# Feature 1 - Number of participating objects of course per week
objs = {}
for i in range(req_len):
    ob = req_train.loc[i, 'object']
    eid = req_train.loc[i, 'enrollment_id']
    index = enrollment_id1.index(eid)
    if eid not in objs:
        objs[eid] = [ob]
        features[index][3] += 1
    else:
        if ob not in objs[eid]:
            objs[eid].append(ob)
            features[index][3] += 1
```

```python
# Feature 2-8 - Behaviour number of access, page_close, problem video, discussion, navigate, wiki
for i in range(req_len):
    eid = req_train.loc[i, 'enrollment_id']
    event = req_train.loc[i, 'event']
    index = enrollment_id1.index(eid)
    if event == 'access':
        features[index][4] += 1
    elif event == 'page_close':
        features[index][5] += 1
    elif event == 'problem':
        features[index][6] += 1
    elif event == 'video':
        features[index][7] += 1
    elif event == 'discussion':
        features[index][8] += 1
    elif event == 'navigate':
        features[index][9] += 1
    else:
        features[index][10] += 1  # wiki
```

```python
# Feature 9-10 - The total and average numbers of all behaviors per week
for i in range(unique_enroll):
    behaviour_list = features[i][4:11]
    features[i][11] = sum(behaviour_list) # Total behaviours per week
    features[i][12] = features[i][11]/7   # Average behaviours per week
```

```python
# Feature 11 - Number of active days per week
active_dates = {}
for i in range(req_len):
    d = req_train.loc[i, 'date']
    eid = req_train.loc[i, 'enrollment_id']
    index = enrollment_id1.index(eid)
    if eid not in active_dates:
        active_dates[eid] = [d]
        features[index][13] += 1
    else:
        if d not in active_dates[eid]:
            active_dates[eid].append(d)
            features[index][13] += 1
```

```python
# Feature 12 - Time consumption per week
t1 = pd.to_datetime(req_train.loc[0, 'time'], infer_datetime_format=True)
e1 = req_train.loc[0, 'enrollment_id']
d1 = req_train.loc[0, 'date']
for i in range(1, req_len):
    t2 = pd.to_datetime(req_train.loc[i, 'time'], infer_datetime_format=True)
    e2 = req_train.loc[1, 'enrollment_id']
    d2 = req_train.loc[1, 'date']
    index = enrollment_id1.index(e2)
    if e1==e2 and d1==d2:
        delta = (t2 - t1).total_seconds()
        features[index][14] += delta
    t1 = t2
    e1 = e2
    d1 = d2    
```

```python
# Feature 13-16 - The behavior numbers of access, page close, problem, video from browser per week respectively.
for i in range(req_len):
    eid = req_train.loc[i, 'enrollment_id']
    source = req_train.loc[i, 'source']
    event = req_train.loc[i, 'event']
    index = enrollment_id1.index(eid)
    if source=='browser':
        if event == 'access':
            features[index][15] += 1
        elif event == 'page_close':
            features[index][16] += 1
        elif event == 'problem':
            features[index][17] += 1
        elif event == 'video':
            features[index][18] += 1
```

```python
# Feature 17-21 - The behavior numbers of access, discussion, problem, navigating, wiki from server per week respectively
for i in range(req_len):
    eid = req_train.loc[i, 'enrollment_id']
    source = req_train.loc[i, 'source']
    event = req_train.loc[i, 'event']
    index = enrollment_id1.index(eid)
    if source=='server':
        if event == 'access':
            features[index][19] += 1
        elif event == 'discussion':
            features[index][20] += 1
        elif event == 'problem':
            features[index][21] += 1
        elif event == 'navigate':
            features[index][22] += 1
        elif event == 'wiki':
            features[index][23] += 1  # wiki
```

```python
# Feature 22-23 - The numbers of all behaviors from browser and server per week respectively
for i in range(unique_enroll):
    browser_list = features[i][15:19]
    features[i][24] = sum(browser_list) # Total behaviours from browser per week
    server_list = features[i][19:24]
    features[i][25] = sum(server_list) # Total behaviours from server per week
```

```python
cols = ['enrollment_id',
        'username',
        'course_id',
        'no_of_objects',
        'bno_access',
        'bno_pageclose',
        'bno_problem',
        'bno_video',
        'bno_discussion',
        'bno_navigate',
        'bno_wiki',
        'bno_total',
        'bno_avg',
        'no_of_active_days',
        'time_consumption',
        'browser_access',
        'browser_pageclose',
        'browser_problem',
        'browser_video',
        'server_access',
        'server_discussion',
        'server_problem',
        'server_navigate',
        'server_wiki',
        'bno_browser',
        'bno_server',
        'label']
        
features_combined = pd.DataFrame(features, columns = cols)
```
