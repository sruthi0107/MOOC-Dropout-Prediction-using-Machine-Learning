```python
import pandas as pd
import os
import numpy as np
import datetime
from datetime import date
```

```python
truth_test = pd.read_csv("test/truth_test.csv")
truth_enroll = truth_test['2'].to_list()
```

```python
required_test_users = pd.read_csv("required_test_users.csv")
required_test_users.drop('Unnamed: 0', axis=1, inplace=True)
#required_test_users = required_test_users.head(10000) # Remove when running full dataset
```

```python
required_test_users
```

```python
enrollment_id_list = required_test_users['enrollment_id'].to_list()
```

```python
week4 = pd.read_csv('week4_test.csv')
week4
```

```python
week4.drop('Unnamed: 0', axis=1, inplace=True)
week4
```

```python
features4 = []
enrollment_id1 = []
week4_len = len(week4)
for i in range(week4_len):
    e_id = week4.loc[i, 'enrollment_id']
    if e_id not in enrollment_id1:
        enrollment_id1.append(e_id)
#enrollment_id1 = enrollment_id1[:2]
for eid in enrollment_id1:
    i = enrollment_id_list.index(eid)
    truth_index = truth_enroll.index(eid)
    temp4 = [eid, required_test_users.loc[i, 'username'], required_test_users.loc[i, 'course_id']]+[0]*23+[truth_test.loc[truth_index, '0']]
    features4.append(temp4)
unique_enroll = len(enrollment_id1)
```

```python
unique_enroll
```

```python
# Feature 1 - Number of participating objects of course per week
objs = {}
for i in range(week4_len):
    ob = week4.loc[i, 'object']
    eid = week4.loc[i, 'enrollment_id']
    index = enrollment_id1.index(eid)
    if eid not in objs:
        objs[eid] = [ob]
        features4[index][3] += 1
    else:
        if ob not in objs[eid]:
            objs[eid].append(ob)
            features4[index][3] += 1
```

```python
# Feature 2-8 - Behaviour number of access, page_close, problem video, discussion, navigate, wiki
for i in range(week4_len):
    eid = week4.loc[i, 'enrollment_id']
    event = week4.loc[i, 'event']
    index = enrollment_id1.index(eid)
    if event == 'access':
        features4[index][4] += 1
    elif event == 'page_close':
        features4[index][5] += 1
    elif event == 'problem':
        features4[index][6] += 1
    elif event == 'video':
        features4[index][7] += 1
    elif event == 'discussion':
        features4[index][8] += 1
    elif event == 'navigate':
        features4[index][9] += 1
    else:
        features4[index][10] += 1  # wiki
```

```python
# Feature 9-10 - The total and average numbers of all behaviors per week
for i in range(unique_enroll):
    behaviour_list = features4[i][4:11]
    features4[i][11] = sum(behaviour_list) # Total behaviours per week
    features4[i][12] = features4[i][11]/7   # Average behaviours per week
```

```python
# Feature 11 - Number of active days per week
active_dates = {}
for i in range(week4_len):
    d = week4.loc[i, 'date']
    eid = week4.loc[i, 'enrollment_id']
    index = enrollment_id1.index(eid)
    if eid not in active_dates:
        active_dates[eid] = [d]
        features4[index][13] += 1
    else:
        if d not in active_dates[eid]:
            active_dates[eid].append(d)
            features4[index][13] += 1
```

```python
# Feature 12 - Time consumption per week
t1 = pd.to_datetime(week4.loc[0, 'time'], infer_datetime_format=True)
e1 = week4.loc[0, 'enrollment_id']
d1 = week4.loc[0, 'date']
for i in range(1, week4_len):
    t2 = pd.to_datetime(week4.loc[i, 'time'], infer_datetime_format=True)
    e2 = week4.loc[1, 'enrollment_id']
    d2 = week4.loc[1, 'date']
    index = enrollment_id1.index(e2)
    if e1==e2 and d1==d2:
        delta = (t2 - t1).total_seconds()
        features4[index][14] += delta
    t1 = t2
    e1 = e2
    d1 = d2    
```

```python
# Feature 13-16 - The behavior numbers of access, page close, problem, video from browser per week respectively.
for i in range(week4_len):
    eid = week4.loc[i, 'enrollment_id']
    source = week4.loc[i, 'source']
    event = week4.loc[i, 'event']
    index = enrollment_id1.index(eid)
    if source=='browser':
        if event == 'access':
            features4[index][15] += 1
        elif event == 'page_close':
            features4[index][16] += 1
        elif event == 'problem':
            features4[index][17] += 1
        elif event == 'video':
            features4[index][18] += 1
```

```python
# Feature 17-21 - The behavior numbers of access, discussion, problem, navigating, wiki from server per week respectively
for i in range(week4_len):
    eid = week4.loc[i, 'enrollment_id']
    source = week4.loc[i, 'source']
    event = week4.loc[i, 'event']
    index = enrollment_id1.index(eid)
    if source=='server':
        if event == 'access':
            features4[index][19] += 1
        elif event == 'discussion':
            features4[index][20] += 1
        elif event == 'problem':
            features4[index][21] += 1
        elif event == 'navigate':
            features4[index][22] += 1
        elif event == 'wiki':
            features4[index][23] += 1  # wiki
```

```python
# Feature 22-23 - The numbers of all behaviors from browser and server per week respectively
for i in range(unique_enroll):
    browser_list = features4[i][15:19]
    features4[i][24] = sum(browser_list) # Total behaviours from browser per week
    server_list = features4[i][19:24]
    features4[i][25] = sum(server_list) # Total behaviours from server per week
```

```python
cols = ['enrollment_id',
        'username',
        'course_id',
        'no_of_objects4',
        'bno_access4',
        'bno_pageclose4',
        'bno_problem4',
        'bno_video4',
        'bno_discussion4',
        'bno_navigate4',
        'bno_wiki4',
        'bno_total4',
        'bno_avg4',
        'no_of_active_days4',
        'time_consumption4',
        'browser_access4',
        'browser_pageclose4',
        'browser_problem4',
        'browser_video4',
        'server_access4',
        'server_discussion4',
        'server_problem4',
        'server_navigate4',
        'server_wiki4',
        'bno_browser4',
        'bno_server4',
        'label']
        
features_week4_test = pd.DataFrame(features4, columns = cols)
```
