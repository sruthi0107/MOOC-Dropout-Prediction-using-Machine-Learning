```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
w1 = pd.read_csv("features_week1.csv")
w2 = pd.read_csv("features_week2.csv")
w3 = pd.read_csv("features_week3.csv")
w4 = pd.read_csv("features_week4.csv")
```

```python
w1.drop('Unnamed: 0', axis=1, inplace=True)
w2.drop('Unnamed: 0', axis=1, inplace=True)
w3.drop('Unnamed: 0', axis=1, inplace=True)
w4.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
print(len(w1))
print(len(w2))
print(len(w3))
print(len(w4))
```

#### W1 and W2 merge

```python
w1w2 = pd.merge(w1, w2, how='outer', on=['enrollment_id', 'username', 'course_id', 'label'] )
w1w2.replace(np.nan, 0, inplace=True)
w1w2
```

```python
w1w2.to_csv('features_w1w2.csv')
```

#### W1, W2 and W3 merge

```python
w1w2w3 = pd.merge(w1w2, w3, how='outer', on=['enrollment_id', 'username', 'course_id', 'label'] )
w1w2w3.replace(np.nan, 0, inplace=True)
w1w2w3
```

```python
w1w2w3.to_csv('features_w1w2w3.csv')
```

#### W1, W2, W3 and W4 merge

```python
w1w2w3w4 = pd.merge(w1w2w3, w4, how='outer', on=['enrollment_id', 'username', 'course_id', 'label'] )
w1w2w3w4.replace(np.nan, 0, inplace=True)
w1w2w3w4
```

```python
w1w2w3w4.to_csv('features_w1w2w3w4.csv')
```
