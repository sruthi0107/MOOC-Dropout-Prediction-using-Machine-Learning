```python
start_end_dates = pd.read_csv("start_end_dates.csv")
start_end_dates
```

```python
start_end_dates.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
from datetime import date
for i in range(len(start_end_dates)):
    d1 = start_end_dates.loc[i,'week1_start'].split('-')
    start_end_dates.at[i,'week1_start'] = date(int(d1[0]), int(d1[1]), int(d1[2]))
    d2 = start_end_dates.loc[i,'week1_end'].split('-')
    start_end_dates.at[i,'week1_end'] = date(int(d2[0]), int(d2[1]), int(d2[2]))
    d3 = start_end_dates.loc[i,'week2_start'].split('-')
    start_end_dates.at[i,'week2_start'] = date(int(d3[0]), int(d3[1]), int(d3[2]))
    d4 = start_end_dates.loc[i,'week2_end'].split('-')
    start_end_dates.at[i,'week2_end'] = date(int(d4[0]), int(d4[1]), int(d4[2]))
    d5 = start_end_dates.loc[i,'week3_start'].split('-')
    start_end_dates.at[i,'week3_start'] = date(int(d5[0]), int(d5[1]), int(d5[2]))
    d6 = start_end_dates.loc[i,'week3_end'].split('-')
    start_end_dates.at[i,'week3_end'] = date(int(d6[0]), int(d6[1]), int(d6[2]))
    d7 = start_end_dates.loc[i,'week4_start'].split('-')
    start_end_dates.at[i,'week4_start'] = date(int(d7[0]), int(d7[1]), int(d7[2]))
    d8 = start_end_dates.loc[i,'week4_end'].split('-')
    start_end_dates.at[i,'week4_end'] = date(int(d8[0]), int(d8[1]), int(d8[2]))
```

```python
required_train_users = pd.read_csv("required_users.csv")
required_train_users.drop('Unnamed: 0', axis=1, inplace=True)
#required_train_users = required_train_users.head(10000)
```

```python
required_train_users
```

```python
enrollment_id_list = required_train_users['enrollment_id'].to_list()
```

```python
new_log_train = pd.read_csv('new_log_train.csv')
new_log_train.drop('Unnamed: 0', axis=1, inplace=True)
#new_log_train = new_log_train.head(10000) # Remove when running full dataset
```

```python
new_log_train
```

```python
courses = datefile.head(10)
course_id_list = courses['course_id'].to_list()
course_id_list
```

```python
week1, week2, week3, week4 = [], [], [], []
```

```python
n = len(new_log_train)
for i in range(n):
    datestr = new_log_train.loc[i, 'date'].split('-')
    d = date(int(datestr[0]), int(datestr[1]), int(datestr[2]))
    e = new_log_train.loc[i, 'enrollment_id']
    if e in enrollment_id_list:
        j = enrollment_id_list.index(e)
        cid = required_train_users.loc[j, 'course_id']
        k = course_id_list.index(cid)
        row = new_log_train.iloc[i]
        row['date'] = d
        if start_end_dates.loc[k, 'week1_start']<=d<=start_end_dates.loc[k, 'week1_end']:
            week1.append(row)
        elif start_end_dates.loc[k, 'week2_start']<=d<=start_end_dates.loc[k, 'week2_end']:
            week2.append(row)
        elif start_end_dates.loc[k, 'week3_start']<=d<=start_end_dates.loc[k, 'week3_end']:
            week3.append(row)
        else:
            week4.append(row)
```

```python
week1 = pd.DataFrame(week1)
week2 = pd.DataFrame(week2)
week3 = pd.DataFrame(week3)
week4 = pd.DataFrame(week4)
```
